```python
from docx import Document

doc = Document('خطة درسية للارشاد (1).docx')
table = doc.tables[1]
table
```

```python
data = []
for row in table.rows:
    row_data = []
    for cell in row.cells:
        row_data.append(cell.text)
    data.append(row_data)
```
